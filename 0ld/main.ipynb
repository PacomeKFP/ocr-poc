```python
import numpy as np
# from paddleocr import PaddleOCR
from collections import Counter
from typing import Dict


FRONT_USELESS_ITEMS = [
    "REPUBLIQUE DU CAMEROUN",
    "NATIONAL IDENITY CARD",
    "REPUBLIC OF CAMEROON",
    "CARTE NATIONALE DIDENTITE",
    "CARTE NATIONALE D'IDENTITE",
    ""
]

FRONT_BOUNDS = [
    "NOM/SURNAME",
    "PRENOMS/GIVEN NAMES",
    "DATE DE NAISSANCE/DATEOF BIRTH",
    "LIEU DE NAISSANCE/PLACE OF BIRTH",
    "SEXE/SEX",
    "TAILLE/HEIGHT",
    "PROFESSION/OCCUPATION",
    "SIGNATURE",
]

BACK_USELESS_ITEMS = [
    "AUTORITE/AUTHORITY",
    "DATEOFISSUE", #repair
    "IDENTIFICATION POS", #repair
    "DATE OFEXPIRY",# repair
    "URIOURIDENTIFIER",# repair
    "CAM動UN",
    "CAMERO",
    ""
]
BACK_BOUNDS = [
    "PERE/FATHER",
    "MERE/MOTHER",
    "S.P./S.M.",
    "DATEDEDELIVRANCE/",
    "POSTEDIDENTIFICATCN",
    "ADRESSE/ADDRESS",
    "DATED'EXPIRATION/",
    "Martin MBARGA NGUELE", # repair
    "IDENTIFIANT UNIQUE",
]


COMPARISON_TRHESHOLD = 0.8



class Extractor:

    def __init__(self):

        self.result = None
        self.ocr_result = None
        self.COMPARISON_TRHESHOLD = 0.8
        self.BOUNDS = []
        self.USELESS_ITEMS = []
        self.front_dict = ["name", "given_names", "birth_date", "birth_place", ]



    def execute(self, ocr_result:Dict, bounds, useless_items):
        self.BOUNDS = bounds
        self.USELESS_ITEMS = useless_items

        # self._execute_orc(image_path)
        self._remove_useless_text(self.ocr_result[0])
        self._remove_useless_text(self.ocr_result[0])
        # self._remove_items_with_minoritary_text_orientation(self.ocr_result[0])
        self._remove_elements_with_score_less_than_average(self.ocr_result[0])
        print(self.ocr_result[0]["rec_texts"])
        self._extract_text(self.ocr_result[0])


        return self._return_result()


    def _execute_orc(self, image_path):

        self.ocr_result = self.ocr.predict(image_path)


    def _remove_items_with_minoritary_text_orientation(self, ocr_result):

        # Retire les items qui sont dans la liste EXCLUDE

        angles = ocr_result["textline_orientation_angles"]

        majoritaire = Counter(angles).most_common(1)[0][0]


        indices = [i for i, v in enumerate(angles) if v == majoritaire]


        ocr_result["textline_orientation_angles"] = [angles[i]
                                                     for i in indices]

        ocr_result["rec_texts"] = [ocr_result["rec_texts"][i] for i in indices]

        ocr_result["rec_scores"] = [ocr_result["rec_scores"][i]
                                    for i in indices]

        ocr_result["rec_polys"] = [ocr_result["rec_polys"][i] for i in indices]

        ocr_result["rec_boxes"] = [ocr_result["rec_boxes"][i] for i in indices]


    def _remove_useless_text(self, ocr_result):
        """

        Retire les textes qui ne sont pas dans la liste BOUNDS
        """

        indices = set()

        for element in self.USELESS_ITEMS:

            for i in range(len(ocr_result["rec_texts"])):

                if StringComparator.compare_strings(ocr_result["rec_texts"][i], element) > self.COMPARISON_TRHESHOLD:

                    indices.add(i)


        indices = set(range(len(ocr_result["rec_texts"]))) - indices


        ocr_result["textline_orientation_angles"] = [

            ocr_result["textline_orientation_angles"][i] for i in indices]

        ocr_result["rec_texts"] = [ocr_result["rec_texts"][i] for i in indices]

        ocr_result["rec_scores"] = [ocr_result["rec_scores"][i]
                                    for i in indices]

        ocr_result["rec_polys"] = [ocr_result["rec_polys"][i] for i in indices]

        ocr_result["rec_boxes"] = [ocr_result["rec_boxes"][i] for i in indices]


    def _remove_elements_with_score_less_than_average(self, ocr_result):
        """

        Retire les elements dont le score est inferieur a la moyenne des scores
        """

        average_score = np.mean(ocr_result["rec_scores"])*0.6

        indices = [i for i, score in enumerate(

            ocr_result["rec_scores"]) if score >= average_score]


        ocr_result["textline_orientation_angles"] = [

            ocr_result["textline_orientation_angles"][i] for i in indices]

        ocr_result["rec_texts"] = [ocr_result["rec_texts"][i] for i in indices]

        ocr_result["rec_scores"] = [ocr_result["rec_scores"][i]
                                    for i in indices]

        ocr_result["rec_polys"] = [ocr_result["rec_polys"][i] for i in indices]

        ocr_result["rec_boxes"] = [ocr_result["rec_boxes"][i] for i in indices]


    def _extract_text(self, ocr_result):

        self.result = dict()

        bounds_indices = []

        for bound in self.BOUNDS:

            indice, score = None, None

            for i in range(len(ocr_result["rec_texts"])):

                similarity_score = StringComparator.compare_strings(

                    ocr_result["rec_texts"][i], bound)

                if similarity_score > self.COMPARISON_TRHESHOLD:

                    if score is None or similarity_score > score:

                        score = similarity_score

                        indice = i

            bounds_indices.append(indice)


        cleaned_bounds_indices = [i for i in bounds_indices if i is not None]

        idx = 0

        while idx < len(cleaned_bounds_indices) - 1:

            bound_idx = cleaned_bounds_indices[idx]

            next_bound_idx = cleaned_bounds_indices[idx + 1]
            bounds_batch = [bound_idx]
            while next_bound_idx is not None and next_bound_idx == bound_idx + 1:
                bounds_batch.append(next_bound_idx)
                bound_idx = next_bound_idx
                next_bound_idx = cleaned_bounds_indices[idx +
                                                        1 + len(bounds_batch) - 1]

            if len(bounds_batch) < 2:
                if bound_idx is not None and next_bound_idx is not None:

                    self.result[ocr_result["rec_texts"][bound_idx]] = {

                        "text": " ".join(ocr_result["rec_texts"][bound_idx + 1: next_bound_idx]),

                        "score": np.mean(ocr_result["rec_scores"][bound_idx + 1: next_bound_idx]).__float__(),

                        "box(es)": ocr_result["rec_boxes"][bound_idx + 1: next_bound_idx],

                        "poly(s)": ocr_result["rec_polys"][bound_idx + 1: next_bound_idx],

                    }

            else:
                # regarder les {len(bounds_batch)} qui viennent apres les bounds_batch[-1]
                # s'il sont tous des non bounds, alors, on map 1 -- 1
                can_go_ahead = True
                # for b in range(bounds_batch[-1], bounds_batch[-1] + len(bounds_batch) + 1):
                #     if b in cleaned_bounds_indices:
                #         can_go_ahead = False
                #         break
                if can_go_ahead:
                    for _bound_idx in bounds_batch:
                        self.result[ocr_result["rec_texts"][_bound_idx]] = {
                            "text": ocr_result["rec_texts"][_bound_idx+len(bounds_batch)],
                            "score": ocr_result["rec_scores"][_bound_idx+len(bounds_batch)].__float__(),
                            "box(es)": ocr_result["rec_boxes"][_bound_idx+len(bounds_batch)],
                            "poly(s)": ocr_result["rec_polys"][_bound_idx+len(bounds_batch)],
                        }

            idx += len(bounds_batch)


    def _return_result(self, include_all_details=False):

        if self.result is None:

            raise ValueError(
                "No result to return. Please execute the OCR first.")


        if include_all_details:

            return self.result

        else:

            simplified_result = {
                key: (value["text"], value["score"]) for key, value in self.result.items()}

            return simplified_result



class StringComparator:

    @staticmethod

    def compare_strings(string1: str, string2: str) -> float:
        """

        Compare two strings and return a similarity score.
        """

        if not string1 or not string2:

            return 0.0

        string1 = string1.lower().replace(" ", "")

        string2 = string2.lower().replace(" ", "")


        max_length = max(len(string1), len(string2))


        if max_length == 0:

            return 1.0


        matching_chars = sum(1 for a, b in zip(string1, string2) if a == b)


        similarity_score = matching_chars / max_length

        return similarity_score
    
```
